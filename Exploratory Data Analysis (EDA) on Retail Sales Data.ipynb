```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline

```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\retail_sales.xlsx.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.fillna(0, inplace=True)   
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Calories'].mean().plot(kind='bar', title='Average Calories by Category')
plt.ylabel("Calories")
plt.show()
```

```python
top_cal_items = df[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
top_cal_items
```

```python
plt.scatter(df['Total Fat'], df['Calories'])
plt.xlabel("Total Fat")
plt.ylabel("Calories")
plt.title("Total Fat vs Calories")
plt.show()
```

```python
df.groupby('Category')['Sodium'].mean().plot(kind='bar', title='Average Sodium by Category')
plt.ylabel("Sodium (mg)")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
